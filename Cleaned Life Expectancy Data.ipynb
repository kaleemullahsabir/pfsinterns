```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\DELL\AppData\Local\Programs\Python\Python312\Task 1 Health and Lifespan Analysis\Life Expectancy Data.csv"
df = pd.read_csv(file_path)

# Display first few rows
df.head()

# Get basic info
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# Fill missing numerical values with the mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Fill missing categorical values with mode
df['Status'] = df['Status'].fillna(df['Status'].mode()[0])

# Convert 'Year' column to DateTime format
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# Verify data after cleaning
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove any extra spaces from column names
df.columns = df.columns.str.strip()

# Plot the distribution of Life Expectancy
plt.figure(figsize=(8, 5))
sns.histplot(df['Life expectancy'], bins=30, kde=True, color='blue')
plt.title("Distribution of Life Expectancy")
plt.xlabel("Life Expectancy (Years)")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove any extra spaces from column names
df.columns = df.columns.str.strip()

# Scatter Plot: Life Expectancy vs Healthcare Expenditure
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Total expenditure'], y=df['Life expectancy'], alpha=0.6, color='blue')
plt.title("Life Expectancy vs Healthcare Expenditure")
plt.xlabel("Healthcare Expenditure (% of GDP)")
plt.ylabel("Life Expectancy (Years)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Compute the correlation matrix
correlation_matrix = df.corr(numeric_only=True)  # Ensures only numerical columns are used

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Regression Plot: Healthcare Spending vs Life Expectancy
plt.figure(figsize=(8, 5))
sns.regplot(x=df['Total expenditure'], y=df['Life expectancy'], scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
plt.title("Relationship Between Healthcare Spending and Life Expectancy")
plt.xlabel("Healthcare Spending (% of GDP)")
plt.ylabel("Life Expectancy (Years)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Regression Plot: Adult Mortality vs Life Expectancy
plt.figure(figsize=(8, 5))
sns.regplot(x=df['Adult Mortality'], y=df['Life expectancy'], scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.title("Life Expectancy vs Adult Mortality")
plt.xlabel("Adult Mortality Rate")
plt.ylabel("Life Expectancy (Years)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Regression Plot: Schooling vs Life Expectancy
plt.figure(figsize=(8, 5))
sns.regplot(x=df['Schooling'], y=df['Life expectancy'], scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.title("Life Expectancy vs Schooling")
plt.xlabel("Average Years of Schooling")
plt.ylabel("Life Expectancy (Years)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Box Plot: Life Expectancy in Developed vs Developing Countries
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Status'], y=df['Life expectancy'])
plt.title("Life Expectancy: Developed vs Developing Countries")
plt.xlabel("Country Status")
plt.ylabel("Life Expectancy (Years)")
plt.show()
```

```python
# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Top 10 Countries with Highest Life Expectancy
top_10 = df.groupby('Country')['Life expectancy'].mean().nlargest(10)
print("Top 10 Countries with Highest Life Expectancy:")
print(top_10)

# Bottom 10 Countries with Lowest Life Expectancy
bottom_10 = df.groupby('Country')['Life expectancy'].mean().nsmallest(10)
print("\nBottom 10 Countries with Lowest Life Expectancy:")
print(bottom_10)
```

```python
import matplotlib.pyplot as plt

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

plt.figure(figsize=(10, 5))

# Choose specific countries
countries = ['India', 'United States of America', 'Nigeria']

for country in countries:
    country_data = df[df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Life expectancy'], label=country, marker='o')

plt.title("Life Expectancy Trends Over Time")
plt.xlabel("Year")
plt.ylabel("Life Expectancy (Years)")
plt.legend()
plt.grid(True)
plt.show()
```
